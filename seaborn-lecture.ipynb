#### Types of data:

- Continuous Data:- Numeric data with possibly infinite resolution
    - Height
    - Weight
    - total_bill and tip
    - temperature
- Categorical - Distinct categories
    - weekdays
    - Gender
    - 'smoker'
    - letter grades (A, B, C..)

- Discrete - Distinct numeric categories 
     - party_size - numeric but discete
     - number of customer complaints
     - number of flaws or defects.

### Relational Plots 

```python
# Explore relationship between total_bill and tip

sns.relplot(data = tips, x = 'total_bill', y = 'tip')

# returns a FacetGrid (which means a figure level object)
```

```python
# plt.style.use()
# Set styling in seaborn
sns.set()

sns.relplot(data = tips, x = 'total_bill', y = 'tip')

plt.title("Total Bill vs Tip Amount")
```

```python
# visualize if dist is different for smoker vs non-smoker
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue='smoker')

```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue='time')

```

```python
# Style as a kwarg
sns.relplot(data = tips, x = 'total_bill', y = 'tip', style='time')
```

```python
# use arg col
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'time', col = 'day')

# use col_wrap to specify number of columns
```

#### Key takeaways?
- Tip amount generally increases with total_bill
- Waiter works in evenings on Sun and sat
- Waiter works during lunchtime on Thursday


#### Ways to add DF columns to the chart
- hue
- col - creates subplots
- style


#### Regression plots

```python
# Add Regression line to relplot

sns.lmplot(x = 'total_bill', y = 'tip', col = 'smoker', data = tips)

# Translucent band is 95% CI. 
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips, ci = 95)

```

### Categorial Plots

#### Boxplots

```python
sns.boxplot(data = tips, x = 'tip')
plt.grid(True, ls = '--')
plt.xticks(np.arange(0,10,1));
```

```python
tips.tip.describe()
```

```python
plt.figure(figsize = (10,8))

plt.subplot(221)
sns.boxplot(y = 'tip', data = tips)

plt.subplot(222)
sns.boxplot(x = 'tip', data = tips)

plt.subplot(223)
sns.boxplot(y = 'tip', x = 'size', data = tips)

plt.subplot(224)
sns.boxplot(x = 'tip', data = tips, y = 'time', hue = 'smoker',
           order = ['Lunch', 'Dinner'], hue_order = ['Yes', 'No'])
```

#### Heatmap

```python
data = pd.crosstab(tips.time, tips['size'])
data
```

```python
sns.heatmap(data, cmap = plt.cm.Greens, annot =True, fmt = 'd')
```

```python
data = pd.crosstab(tips.time, tips.smoker)
data
```

```python
sns.heatmap(data, annot = True, fmt = 'd')
```

#### Pairplot

```python
sns.pairplot(tips, corner = True)
```

```python
sns.pairplot(tips, hue = 'smoker')
```

#### Jointplot

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'reg')
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hex')
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde', fill = True)
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', 
              kind = 'kde', cmap = "magma", fill = True
             , thresh = 0, levels = 60)
```
